```sql
SELECT ai_analyze_sentiment('I am happy')
```

```sql
select *,ai_analyze_sentiment(review) as sentiment  from samples.bakehouse.media_customer_reviews
```

```sql
create temp view analyse_sentiment as 
select *,ai_analyze_sentiment(review) as sentiment  from samples.bakehouse.media_customer_reviews
```

```sql
select sentiment, count(sentiment) from analyse_sentiment group by sentiment 
```

```sql
SELECT ai_classify("My password is leaked.", ARRAY("urgent", "not urgent"))
```
