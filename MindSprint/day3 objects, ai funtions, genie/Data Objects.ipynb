```sql
create or replace function dev_agents.deepak.mask_email(email string)
returns string
return concat(
  left(email, 1),
  repeat('*', length(split(email, '@')[0]) - 1),
  '@',
  split(email, '@')[1]
);
```

```sql
select dev_agents.naval.mask_email('naval@gmail.com') as email
```

```sql
create or replace function dev_agents.naval.full_name(fname string, lname string)
returns string
return concat(fname, ' ', lname)
```

```sql
select dev_agents.naval.full_name('Sanju','Samson') as name
```
